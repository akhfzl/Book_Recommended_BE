```python
saintech = ['analys', 'herbal','relativity','develop','robot','Extrasensory','Psycho', 'Disease','Medical', 'Aids','Computers', 'Computer', 'Math', 'unix','Technology', 'Math', 'Health', 'Computa', 'Probabil','latex','Physical','computer','Astronomi','radiograph','Math','syndrome','Big data', 'mining', 'Chemistry','Discoveries' ,'Geometry','chemistry','Engineering', 'Cancer', 'geograph','network','Animals','Electronic', 'Physi', 'forensic','Brainwashing', 'clinic','Memori', 'memory', 'climate','Nuclear', 'Alchemy', 'Liquid', 'Techno', 'Distraction', 'astronaut','Archi', 'Database', 'Data', 'programming', 'earth']
soshum = ['money', 'economic','police','Law', 'Business', 'e-commerce', 'e-commercial', 'social', 'history', 'art', 'economic', 'Holocaust', 'Advertis', 'finance', 'market', 'Bankers', 'Politic', 'Physiology','human', 'adapt', 'histori','history', 'relations','literature', 'graphic', 'Biographies', 'Biography', 'Relationship', 'financial', 'cultur', 'war', 'manage','sale', 'seller', 'design', 'government', 'philosoph', 'issue', 'Hieroglyphic', 'translat', 'culture', 'Attorney', 'prosecutor', 'judge']
```

```python
target_class = []
for title in books['title']:
    sains = [i for i in saintech if re.search(r'{}'.format(i), title, flags=re.IGNORECASE)]
    sos = [i for i in soshum if re.search(r'{}'.format(i), title, flags=re.IGNORECASE)]
    if len(sains) > 0 and len(sos) < 1:
        target_class.append('sains-tech')
    elif len(sos) > 0 and len(sains) < 1:
        target_class.append('social-humanaria')
    elif len(sos) < 1 and len(sains) < 1: 
        target_class.append('general')
    else:
        target_class.append('sains-tech')
        
books['target_class'] = target_class
books.head(10)
```

Modeling and Tokenize

```python
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
import json
import pickle5 as pickle
from sklearn.metrics import accuracy_score, f1_score, classification_report
```

```python
le = LabelEncoder()
tfidf = TfidfVectorizer(
    min_df=1,  # min count for relevant vocabulary
    max_features=4000,  # maximum number of features
    strip_accents='unicode',  # replace all accented unicode char 
    # by their corresponding  ASCII char
    analyzer='word',  # features made of words
    token_pattern=r'\w{1,}',  # tokenize only words of 4+ chars
    ngram_range=(1, 1),  # features made of a single tokens
    use_idf=True,  # enable inverse-document-frequency reweighting
    smooth_idf=True,  # prevents zero division for unseen words
    sublinear_tf=False)
```

```python
features = tfidf.fit_transform(new_title)
label = le.fit_transform(df['category'].values)
```

```python
pd.DataFrame(features.toarray(), columns=tfidf.vocabulary_, index=new_title)
```

```python
keyValues = {int(label[i]) :le.inverse_transform(label)[i] for i in range(len(le.inverse_transform(label)))}
keyValues
```

```python
with open('result/labeling.json', 'w', encoding='utf-8') as output:
    dump = json.dumps(keyValues)
    output.write(dump)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features.toarray(), label,test_size=0.33, random_state=42)
```

```python
y_test.shape
```

```python
cosine_x_train = cosine_similarity(x_train)
cosine_x_test = cosine_similarity(x_test, x_train)
```

```python
# cosine_x_test.shape
y_test.shape
```

```python
model = SVC(kernel=cosine_similarity)
model.fit(x_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cosine_x_test.shape
```

```python
scores = cross_val_score(model, X=x_test, y=y_test)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
